```python
import pandas as pd
import numpy as np

data = pd.read_csv('Model/Tool_Prediction.csv')
```

```python
next_time = data['next_time']
predicted_time = data['predicted_next_time']

data['time difference'] = next_time - predicted_time
print(data['time difference'].head())

```

```python

print(data['time difference'].head())
```

```python
#plot data

import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
fig = plt.figure()
ax = plt.axes()
```

```python
data_to_plot = data['time difference'][1000:]
#remove nan values
data_to_plot = data_to_plot.dropna()
#remove outliers from data
#data_to_plot = data_to_plot[data_to_plot.between(data_to_plot.quantile(.5), data_to_plot.quantile(.95))]

print(data_to_plot)

# plot normally
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True
plt.axhline(y=0, color='orange', linestyle='-', zorder=2)
plt.plot(data_to_plot, zorder=1)
plt.legend(['Truth value',], loc='lower right', )

plt.title('Time difference between predicted time and actual time')
plt.xlabel('Entries')
plt.ylabel('Time difference')
plt.show()
#draw a orange line at 0 on top of the plot


```
